```python
#read wav to plot and save after reserving 
import wave
import matplotlib.pyplot as plt
import numpy as np
#read wav
f = wave.open('./data_science_lesson1.wav','rb')
params = f.getparams()
nchannels, sampwidth, framerate, nframes = params[:4]
strData = f.readframes(nframes)
waveData = np.frombuffer(strData,dtype=np.int16)
waveData = waveData*1.0/(max(abs(waveData)))
time = np.arange(0,nframes)*(1.0 / framerate)
plt.plot(time,waveData)
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.show()
#reserve and save it in a new file
import struct
outData=waveData[::-1]
outwav=wave.open("./reverse.wav","w")
fs = 8000
data_size = len(outData)
framerate = int(fs)
nframes = data_size
comptype = "NONE"
compname = "not compressed"
outwav.setparams((nchannels, sampwidth, framerate, nframes,
    comptype, compname))
for v in outData:
        outwav.writeframes(struct.pack('h', int(v * 64000 / 2)))
outwav.close()



#read jpg and save after converting
import matplotlib.image as mpimg

img = mpimg.imread('./data_science_lesson1.jpg')
img.shape
plt.imshow(img)
plt.axis('off')
plt.show()
#convert
plt.imshow(img[:,:,1],cmap="nipy_spectral")
plt.colorbar()
plt.savefig('./repaint.jpg')

#read csv and save atfer sorting by scores
import pandas
fdata=pandas.read_csv('./data_science_lesson1.csv',index_col="score")
print(fdata)
fdata.to_csv('./sort.csv')
```
